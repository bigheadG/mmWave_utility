```python
from scipy.fftpack import fft
import numpy as np
from matplotlib import pyplot as plt 
 
v = [8,8,8,1,8,2,4,8,5,8,8,8,8,8,8]

def parkingBins(v4d = None):
    len4 = len(v4d)
    if len4 != 0:
        v4d = np.fft.fftshift(v4d)
        v4d = np.append(v4d,v4d[0])
        x1 = np.linspace(-1,1,len4+2)
        y1 = np.sqrt(1 - x1 * x1)
        
        plt.plot(x1,y1) 
        plt.show()
        
        x_1 = v4d * x1[0:-1]
        x_2 = v4d * x1[1:]
        y_1 = v4d * y1[0:-1]
        y_2 = v4d * y1[1:]
        xa = []
        ya = []
        for i in range(len(x_1)):
            xa.append(x_1[i])
            xa.append(x_2[i])
            ya.append(y_1[i])
            ya.append(y_2[i])
        xa = xa[1:-1]
        ya = ya[1:-1]
        return xa,ya

    #print(np.array(xa).shape)

xa,ya = parkingBins(v4d = v)
    
plt.plot(xa,ya) 
plt.show()
```
